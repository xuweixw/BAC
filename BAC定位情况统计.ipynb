### 统计BAC基因组覆盖大小和覆盖Gap数目

```python
import intervals
import pandas
```

```python
GFF3 = pandas.read_table("../workspace/panicum/PM-B-BES_20190406.gff3",header=None)
GAP = pandas.read_table("../workspace/panicum/Panicum_miliaceum_longmi4_gap.gff3", header=None)
```

```python
GFF3 = GFF3[GFF3[2] == 'BES']
```

```python
def gap_count(GAP, CHR):
    count = 0
    for i in range(GAP.shape[0]):
        gap = intervals.closed(GAP.iloc[i,3],GAP.iloc[i,4])
        if gap in CHR:
            count += 1
            #print(GAP1.iloc[i,8])
    return count
    
```

```python
for i in range(1, 19): # discuss every chromosome
    chr = GFF3[GFF3[0] == str(i)] # in current chromosome, BAC annotation with two BESs
    #print("chr%2s:\tBAC=%s"% (i, chr.shape[0]))  # how many BACs is in current chromosome
    zero = intervals.empty()
    for j in range(chr.shape[0]):
        a = intervals.closed(int(chr.iloc[j,3]), int(chr.iloc[j,4]))
        zero = zero | a
    
    GAP1 = GAP[GAP[0] == i]
    #print("\tGAP=%s" % GAP1.shape[0])
    count = gap_count(GAP1,zero)
    #print("\tfill_gap=%s" %count)
    print("chr%2s:\t%s\t%s\t%s\t%s" % (i, chr.shape[0], GAP1.shape[0], sum([atom.upper - atom.lower for atom in zero]) / 1000000,count))
```

```python
chr = GFF3[GFF3[0] == str(1)] # in current chromosome, BAC annotation with two BESs
#print("chr%2s:\tBAC=%s"% (i, chr.shape[0]))  # how many BACs is in current chromosome
zero = intervals.empty()
for j in range(chr.shape[0]):
    a = intervals.closed(int(chr.iloc[j,3]), int(chr.iloc[j,4]))
    zero = zero | a
```
