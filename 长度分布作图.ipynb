```python
# 20190425
lenbes["9"] = 0
for i in range(lenbes.shape[0]):
    lenbes.iloc[i,9] = int(lenbes.iloc[i,4]) - int(lenbes.iloc[i,3])
print(lenbes[:3])

mean = np.mean(lenbes["9"])
std= np.std(lenbes["9"])
print(mean,std)





plot = (ggplot(lenbes, aes(x="9")) 
+ geom_histogram(binwidth=2000, fill=c_col[3], color='white') 
#+ scale_x_continuous(breaks=[0,50,100,150,200,250,300,350,400,450,500,550,600,750])
+ xlim([0,200000])
+ xlab('Lengths of BAC clones') + ylab('Number of sequences')
+ annotate('text', x=170000, y=320, label=r'$\mu$ = %s kb   $\delta$ = %s kb'%(int(mean)/1000,int(std)/1000), size=24)
+ theme(figure_size=(20, 10),text=element_text(size=24))
)
```

```python
long_BES = [len(i) for i in HTSeq.FastaReader('../Documents/毕业论文/data/long_BES.fasta')]
```

```python
data = pd.DataFrame({'len':long_BES})
```

```python
(ggplot(data ,aes(x='len')) 
 + geom_histogram(binwidth=2000, fill='blue', color='white')
 + scale_x_continuous(breaks=[25000,50000,75000,100000,125000,150000], 
                      labels=['25 kb','50 kb', '75 kb','100 kb','125 kb','150 kb'],limits=(0,160000))
 + annotate('text', x=100000, y=600, 
            label='NL50     = %s\nmean    = %s\nmedian = %s'%('60.873 kb','32.048 kb', '19.834 kb'), 
            size=8)
 + xlab('Lengths of long BAC end sequences') + ylab('Number of sequences')
 + theme(dpi=170)
)
```
