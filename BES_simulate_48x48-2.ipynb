```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
import HTSeq
```

```python
F_end = HTSeq.FastaReader('./PM-B-BES_F_valid.fasta')
F_len = np.array([len(i) for i in F_end])

R_end = HTSeq.FastaReader('./PM-B-BES_R_valid.fasta')
R_len = np.array([len(i) for i in R_end])





def plot_sub(data, id=''):
    num_bins = range(0,801,10)
    mu = np.mean(data)
    sigma = np.std(data)
    n, bins, patches = plt.hist(data, bins=num_bins,color='g', edgecolor='b')
    plt.axis([0,850,0,400])
    plt.xlabel('Length of BESs (bp)')
    plt.ylabel('No. of BESs')
    plt.title('The distribution of the BAC %s-end sequence length' %id)
    plt.text(560,300,'$\mu=%s bp$ , $\sigma=%s bp$'%(int(mu),int(sigma)))


plt.figure(dpi=400)
plt.subplot(211)
plot_sub(F_len, id='forward')

plt.subplot(212)
plot_sub(R_len, id='reverse')

plt.tight_layout() #plt.show()
plt.savefig('the_distribution_48x48.png')
```

```python
cmap = matplotlib.colors.ListedColormap(['white', 'forestgreen', 'blue', 'red'])
norm = matplotlib.colors.BoundaryNorm([0,1,2,3,4],cmap.N)
```

```python
def plot_heatmap(data):
    im = plt.imshow(data, cmap=cmap, norm=norm)
    ax = plt.gca()

    cbar = ax.figure.colorbar(mappable=im, ax=ax)
    cbar.ax.set_ylabel('No. of BESs', rotation=-90, va='bottom')
    cbar.ax.get_yaxis().set_ticks([])
    for j, lab in enumerate(['$0$','$1$','$2$','$3$']):
        cbar.ax.text(.5, (2 * j + 1) / 8.0, lab, ha='center', va='center')


    #cbar.ax.set_xticks[np.array([0.5, 1.5, 2.5, 3.5, 4.5, 5.5])]
    #cbar.ax.set_xticklabels(np.array(['0','1','2','3','4','5']))

    ax.set_xticks(np.array([7,15,23,31,39,47]))
    ax.set_xticklabels(np.array(['Y08','Y16','Y24','Y32','Y40','Y48']),fontdict={'fontsize':6})

    ax.set_yticks(np.array([7,15,23,31,39,47]))
    ax.set_yticklabels(np.array(['X08','X16','X24','X32','X40','X48']),fontdict={'fontsize':6})

    ax.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)
```

```python
data_F = np.load('./PM-F.npy')[:48,:48]
plt.figure(dpi=600)
plot_heatmap(data_F)
plt.savefig('BES_F.png',dpi=600)
plt.show()
```

```python
data_R = np.load('./R_the_number_of_BESs.npy')[:48,:48]
plt.figure(dpi=600)
plot_heatmap(data_R)
plt.savefig('BES_R.png',dpi=600) 
plt.show()
```
